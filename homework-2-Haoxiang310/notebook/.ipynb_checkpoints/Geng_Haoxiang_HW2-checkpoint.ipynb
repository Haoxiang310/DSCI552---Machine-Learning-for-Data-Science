### (c) Simple Linear Regression

```python
def linear_regression_with_plot(data, predictor, response, method, threshold):

    X = sm.add_constant(data[predictor])
    Y = data[response]
    model_simple_regression = sm.OLS(Y, X).fit()
    
    print(f"Regression results for predictor: {predictor} using {method} to mark out outliers")
    print(model_simple_regression.summary())
    
    if method == 'zscore':
        is_outlier_or_not, outlier_list = zscore_get_outliers(data, predictor, threshold)

    elif method == 'studentized_residuals':
        is_outlier_or_not, outlier_list = studentized_residuals_get_outliers(data, model_simple_regression, threshold)

    sns.regplot(data=data, x=predictor, y=response, scatter=False, color = 'green')
    sns.scatterplot(data=data, x=predictor, y=response, hue=is_outlier_or_not, legend=False)
    plt.title(f"Scatter plot with regression line for {predictor} using {method}")
    plt.show()

    print(f"Outlier list using {method}: ")
    print(outlier_list)

# apply z-score method to get outliers
def zscore_get_outliers(data, column, threshold):
    zscores = (data[column] - data[column].mean()) / data[column].std()
    is_outlier_or_not = (threshold < np.abs(zscores))
    outlier_list = data[is_outlier_or_not]
    return is_outlier_or_not, outlier_list

# apply studentized residuals method to get outliers
def studentized_residuals_get_outliers(data, model, threshold):
    studentized_residuals = model.get_influence().resid_studentized_external
    is_outlier_or_not = (threshold < np.abs(studentized_residuals))
    outlier_list = data[is_outlier_or_not]
    return is_outlier_or_not, outlier_list
```

```python
linear_regression_with_plot(data, 'AT','PE', 'zscore', 2)

linear_regression_with_plot(data, 'AT','PE', 'studentized_residuals', 2)
```

```python
linear_regression_with_plot(data, 'V', 'PE', 'zscore', 2)

linear_regression_with_plot(data, 'V', 'PE', 'studentized_residuals', 2)
```

```python
linear_regression_with_plot(data, 'AP', 'PE', 'zscore', 2)

linear_regression_with_plot(data, 'AP', 'PE', 'studentized_residuals', 2)
```

```python
linear_regression_with_plot(data, 'RH', 'PE', 'zscore', 2)

linear_regression_with_plot(data, 'RH', 'PE', 'studentized_residuals', 2)
```

### (f) Nonlinear Association

```python
predictors = ['AT', 'V', 'AP', 'RH']
response = 'PE'

for predictor in predictors:
    poly_predictors = PolynomialFeatures(3)
    poly = poly_predictors.fit_transform(data[predictor].to_numpy().reshape(-1, 1))
    #print(poly)
    poly = sm.add_constant(poly)
    
    model_poly_regression = sm.OLS(data[response], poly).fit()

    print(f"Results of predictor {predictor}:\n",model_poly_regression.summary(),"\n")
```

### Description of Results:

- For predictors AT, AP, RH, each of their p-value x1, x2, x3 are all 0, therefore we can reject null hypothesis for them, and there is significant association between nonlinear form of predictors AT, AP, RH with response PE. 

- For feature V, the p-value for x1 is 0, for x3 is 0.014, which means for predictor V, we can reject the null hypothesis and its x1 and x3 form have association with response PE. But for x2, the p-value is 0.768, which is a lot greater than 0.05, means that we cannot reject null hypothesis and there is no evidence shows that x2 has relationship with response PE.

### (g) Interactions of Predictors

```python
import statsmodels.formula.api as smf

pairwise_interation_terms_formula = "PE ~ AT + V + AP + RH + AT * V + AT * AP + AT * RH + V * AP + V * RH + AP * RH"

model_pairwise_interaction = smf.ols(pairwise_interation_terms_formula, data).fit()

print(model_pairwise_interaction.summary())
```

### Considering when p-value is less than 0.05, there is evidence shows that association exist between predictor and response

- AT:V (p-value = 0.000): This interaction term is statistically significant, indicating that there is an association between the interaction of AT and V with PE.

- AT:AP (p-value = 0.452): This interaction is greater than 0.05 thus there is no evidence of association between the interaction of AT and AP with PE.

- AT:RH (p-value = 0.000): This interaction term is statistically significant, indicating evidence of an association between the interaction of AT and RH with PE.

- V:AP (p-value = 0.000): This interaction term is statistically significant, indicating evidence of an association between the interaction of V and AP with PE.

- V:RH (p-value = 0.086): This interaction is greater than 0.05 thus there is no evidence of association between the interaction of V and RH with PE.

- AP:RH (p-value = 0.034): This interaction term is less than 0.05, indicating evidence of an association between the interaction of AP and RH with PE.

### In conclusion:  the interactions AT:V, AT:RH, V:AP, and AP:RH are statistically significant at the p-value = 0.05 level. This means there's evidence suggesting an association of these interactions with the response variable PE. However, the interactions AT:AP and V:RH are not statistically significant.

### Separate train set and test set:

```python
length = len(data)
data_indices = list(range(length))
random.seed(length)
random.shuffle(data_indices)

train_set_size = int(0.7 * length)

#separate the indices
train_indices = data_indices[0:train_set_size]
test_indices = data_indices[train_set_size:length]

train_set_predictor = data.iloc[train_indices, 0:4] 
train_set_response = data.iloc[train_indices, 4:5]
test_set_predictor = data.iloc[test_indices, 0:4] 
test_set_response = data.iloc[test_indices, 4:5] 

print("train set predictor shape is :", train_set_predictor.shape)
print("train set response shape is :", train_set_response.shape)
print("test set predictor shape is :", test_set_predictor.shape)
print("test set response shape is :", test_set_response.shape)
```

### 1. Baseline model:

```python
baseline_formula = "PE ~ AT + V + AP + RH"

baseline_model = smf.ols(baseline_formula, data).fit()

baseline_train_prediction_results = baseline_model.predict(train_set_predictor)
baseline_test_predictions_results = baseline_model.predict(test_set_predictor)

baseline_train_error = mean_squared_error(train_set_response, baseline_train_prediction_results)
baseline_test_error = mean_squared_error(test_set_response, baseline_test_predictions_results)

print(baseline_model.summary(),"\n")

print("Train error for baseline regression model is:", baseline_train_error,"\n")

print("Test  error for baseline regression model is:", baseline_test_error,"\n")
```

### (i) KNN

```python
scaler = MinMaxScaler()
train_set_predictor_normalized = scaler.fit_transform(train_set_predictor)
test_set_predictor_normalized = scaler.fit_transform(test_set_predictor)

k_values = range(1, 101, 1)

train_error_raw = []
test_error_raw = []

train_error_normalized = []
test_error_normalized = []

for k in k_values:
    knn_model_raw_data = KNeighborsRegressor(n_neighbors=k)
    knn_model_raw_data.fit(train_set_predictor, train_set_response)
    
    train_predict_result_raw = knn_model_raw_data.predict(train_set_predictor)
    test_predict_result_raw = knn_model_raw_data.predict(test_set_predictor)
    train_error_raw.append(mean_squared_error(train_set_response, train_predict_result_raw))
    test_error_raw.append(mean_squared_error(test_set_response, test_predict_result_raw))

    knn_model_normalized_data = KNeighborsRegressor(n_neighbors=k)
    knn_model_normalized_data.fit(train_set_predictor_normalized, train_set_response)
    
    train_predict_result_normalized = knn_model_normalized_data.predict(train_set_predictor_normalized)
    test_predict_result_normalized = knn_model_normalized_data.predict(test_set_predictor_normalized)
    train_error_normalized.append(mean_squared_error(train_set_response, train_predict_result_normalized))
    test_error_normalized.append(mean_squared_error(test_set_response, test_predict_result_normalized))

inverse_k = [1/k for k in k_values]

plt.plot(inverse_k, train_error_raw, label='Train Error')
plt.plot(inverse_k, test_error_raw, label='Test Error')
plt.title('KNN Regression with Raw Features')
plt.xlabel('1/k')
plt.ylabel('Mean Square Error')
plt.legend()
plt.show()

plt.plot(inverse_k, train_error_normalized, label='Train Error')
plt.plot(inverse_k, test_error_normalized, label='Test Error')
plt.title('KNN Regression with Normalized Features')
plt.xlabel('1/k')
plt.ylabel('Mean Square Error')
plt.legend()
plt.show()

#Since when k=1, train error would always be 0. Thus, we need to filter out k=1 when obtaining the minimum train error in both raw data and normalized data.
index_min_train_error_raw = train_error_raw.index(min(train_error_raw[1:]))
index_min_test_error_raw = test_error_raw.index(min(test_error_raw))
index_min_train_error_normalized = train_error_normalized.index(min(train_error_normalized[1:]))
index_min_test_error_normalized = test_error_normalized.index(min(test_error_normalized))

results = {
    "Raw": {
        "Train": {"Error": min(train_error_raw[1:]), "k": k_values[index_min_train_error_raw]},
        "Test": {"Error": min(test_error_raw), "k": k_values[index_min_test_error_raw]}
    },
    "Normalized": {
        "Train": {"Error": min(train_error_normalized[1:]), "k": k_values[index_min_train_error_normalized]},
        "Test": {"Error": min(test_error_normalized), "k": k_values[index_min_test_error_normalized]}
    }
}

print("Results of raw data:", results["Raw"], "\n")
print("Results of normalized data:", results["Normalized"], "\n")
```

### In linear regression, even if we incorporate terms with interactions and up to cubic terms, we still don't take into account terms beyond the third degree and more complex interaction terms. Additionally, we still operate within the linear relations among terms. Thus, if the pattern of the data deviates significantly from linear regression model, we won't achieve good results. In contrast, KNN is more flexible than linear regression, which does not make prior assumptions about the pattern of the data; it classifies based on distance. This might be a reason why linear regression does not perform as well as KNN.

## 2. ISLR: 2.4.1

### (a) The sample size n is extremely large, and the number of predictors p is small.

A flexible statistical learning method would generally perform better.
With a large number of data points and small number of predictors, a inflexible method may not be able to predict accurate coefficients, thus will cause underfit and couldn't obtain the pattern very well.

### (b) The number of predictors p is extremely large, and the number of observations n is small.

An inflexible statistical learning method would generally perform better.
For example, in KNN, if number of observations is small, it will be very likely to cause overfit in knn models. However, more predictors can help the inflexible model to generate more accurate coefficient.

### (c) The relationship between the predictors and response is highly non-linear.

A flexible statistical learning method would generally perform better.
An inflexiblee learning method will be hard to predict the accurate coefficients due to the non-linear pattern of data. For example, in linear regression, if the data is non-linear, we will definity get a relatively bad result rather than flexible methods.

### (d) The variance of the error terms, i.e. $^2$ = Var(), is extremely high.

An inflexible statistical learning method would generally perform better.
High error variance suggests that there's a lot of noise in the data. Flexible models might attempt to fit to this noise, leading to overfitting. On the contrary, a inflexible method may be more robust under a lot of noise.
