### (b) Data Preparation

#### (i) Research what types of techniques are usually used

- Mean/Median/Mode Imputation:    
Mean Imputation: Replace missing values with the mean value of the entire feature column.
Median Imputation: Use the median value for imputation which is robust to outliers.
Mode Imputation: For categorical data, the mode (most frequent value) is often used.
    
- K-Nearest Neighbors (KNN) Imputation:    
Impute missing values based on the K nearest neighbors found in the training set. It considers the feature space to estimate the missing values.
    
- Regression Imputation:    
Utilize a regression model to predict the missing values based on other features.
    
- Interpolation and Extrapolation:    
In time series or spatial data, these methods use patterns to estimate missing values.    

- MICE (Multiple Imputation by Chained Equations):        
A strategy for imputing missing values by modeling each variable with missing values as a function of other variables in a round-robin fashion.    

- Expectation-Maximization (EM) Algorithm:    
A probabilistic approach that estimates the missing values by maximizing the likelihood function

#### (ii) Calculate the coefficient of variation

```python
def split_data(data):
    X = data.iloc[:, 1:]
    Y = data.iloc[:, 0:1]
    return X, Y

aps_train_X, aps_train_Y = split_data(aps_train_data)
aps_test_X, aps_test_Y = split_data(aps_test_data)

print(aps_train_X.shape)
print(aps_train_Y.shape)
print(aps_test_X.shape)
print(aps_test_Y.shape)
```

```python
# Define the imputer to replace missing values with the mean of each column
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

aps_train_X_temp = imputer.fit_transform(aps_train_X)
imputed_aps_train_X = pd.DataFrame(aps_train_X_temp, columns=aps_train_X.columns)
print(imputed_aps_train_X)

# aps_test_X = imputer.transform(aps_test_X)
aps_test_X_temp = imputer.fit_transform(aps_test_X)
imputed_aps_test_X = pd.DataFrame(aps_test_X_temp, columns=aps_test_X.columns)
print(imputed_aps_test_X)
```

```python
means = imputed_aps_train_X.mean()
stds = imputed_aps_train_X.std()
cv = stds / means
result_df = pd.DataFrame([stds, means, cv], 
                        index=['Standard Deviation', 'Mean', 'Coefficient of Variation'])
print(result_df)
```

### (c) Train a random forest

```python
rf = RandomForestClassifier(n_estimators=50, max_depth = 5,oob_score=True, random_state=42)
rf.fit(imputed_aps_train_X, aps_train_Y)
```

```python
train_preds = rf.predict(imputed_aps_train_X)
train_probs = rf.predict_proba(imputed_aps_train_X)[:, 1]
test_preds = rf.predict(imputed_aps_test_X)
test_probs = rf.predict_proba(imputed_aps_test_X)[:, 1]
```

```python
# Confusion matrix
train_confusion_matrix = confusion_matrix(aps_train_Y, train_preds, labels=['neg', 'pos'])
test_confusion_matrix = confusion_matrix(aps_test_Y, test_preds, labels=['neg', 'pos'])

train_fpr, train_tpr, _ = roc_curve(aps_train_Y, train_probs, pos_label='pos')
train_roc_auc = auc(train_fpr, train_tpr)

test_fpr, test_tpr, _ = roc_curve(aps_test_Y, test_probs, pos_label='pos')
test_roc_auc = auc(test_fpr, test_tpr)

#errs
train_err = 1 - accuracy_score(aps_train_Y, train_preds)
test_err = 1 - accuracy_score(aps_test_Y, test_preds)
out_of_bag_err = 1 - rf.oob_score_

print(f"Training Confusion Matrix:\n{train_confusion_matrix}")
print(f"Test Confusion Matrix:\n{test_confusion_matrix}")
print(f"Training ROC AUC: {train_roc_auc}")
print(f"Test ROC AUC: {test_roc_auc}")
print(f"Training Error: {train_err}")
print(f"Test Error: {test_err}")
print(f"OOB Error Estimate: {out_of_bag_err}")

#roc and auc
plt.plot(train_fpr, train_tpr, color='blue', lw=2, label=f'Train ROC curve (area = {train_roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Training Set ROC')
plt.legend(loc="lower right")
plt.show()
plt.plot(test_fpr, test_tpr, color='darkorange', lw=2, label=f'Test ROC curve (area = {test_roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Test Set ROC')
plt.legend(loc="lower right")
plt.show()
```

### (d) Research class imbalance in random forest

- Class Weight Adjustment:
Adjust the class_weight parameter in the RandomForestClassifier, setting it to "balanced" or "balanced_subsample" which automatically weights classes inversely proportional to their frequency.

- Balanced Random Forest Classifier:
A BalancedRandomForestClassifier is a specialized random forest implementation that randomly downsamples each bootstrap sample to balance it.  We may also apply upsampling or use SMOTE to make bootstrap samples balance.

```python
rf_balanced = RandomForestClassifier(n_estimators=50, max_depth = 5, oob_score=True, random_state=42, class_weight='balanced')
rf_balanced.fit(imputed_aps_train_X, aps_train_Y)

train_preds_balanced = rf_balanced.predict(imputed_aps_train_X)
train_probs_balanced = rf_balanced.predict_proba(imputed_aps_train_X)[:, 1]
test_preds_balanced = rf_balanced.predict(imputed_aps_test_X)
test_probs_balanced = rf_balanced.predict_proba(imputed_aps_test_X)[:, 1]

train_confusion_matrix_balanced = confusion_matrix(aps_train_Y, train_preds_balanced, labels=['neg', 'pos'])
test_confusion_matrix_balanced = confusion_matrix(aps_test_Y, test_preds_balanced, labels=['neg', 'pos'])

train_fpr_balanced, train_tpr_balanced, _ = roc_curve(aps_train_Y, train_probs_balanced, pos_label='pos')
train_roc_auc_balanced = auc(train_fpr_balanced, train_tpr_balanced)
test_fpr_balanced, test_tpr_balanced, _ = roc_curve(aps_test_Y, test_probs_balanced, pos_label='pos')
test_roc_auc_balanced = auc(test_fpr_balanced, test_tpr_balanced)

#errs
train_err_balanced = 1 - accuracy_score(aps_train_Y, train_preds_balanced)
test_err_balanced = 1 - accuracy_score(aps_test_Y, test_preds_balanced)
out_of_bag_err_balanced = 1 - rf_balanced.oob_score_

print("Balanced Model Metrics:")
print(f"Training Confusion Matrix:\n{train_confusion_matrix_balanced}")
print(f"Test Confusion Matrix:\n{test_confusion_matrix_balanced}")
print(f"Training ROC AUC: {train_roc_auc_balanced}")
print(f"Test ROC AUC: {test_roc_auc_balanced}")
print(f"Training Misclassification Rate: {train_err_balanced}")
print(f"Test Misclassification Rate: {test_err_balanced}")
print(f"OOB Error Estimate: {out_of_bag_err_balanced}")

#roc and auc
plt.plot(train_fpr_balanced, train_tpr_balanced, color='blue', lw=2, label=f'Train AUC = {train_roc_auc_balanced:.2f}')
plt.plot([0, 1], [0, 1], 'r--',color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.title('Training ROC Curve (Balanced Model)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
plt.plot(test_fpr_balanced, test_tpr_balanced, color='orange', lw=2, label=f'Test AUC = {test_roc_auc_balanced:.2f}')
plt.plot([0, 1], [0, 1], 'r--',color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.title('Test ROC Curve (Balanced Model)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

- Limit both the number of trees and max_depth of trees:

### (e) XGBoost and Model Trees

```python
label_encoder = LabelEncoder()
aps_train_Y_encoded = label_encoder.fit_transform(aps_train_Y)
aps_test_Y_encoded = label_encoder.transform(aps_test_Y)
xgb = XGBClassifier(objective='binary:logistic', learning_rate=0.1, max_depth=5, use_label_encoder=False, eval_metric='logloss')
param_grid = {
    'reg_alpha': [0.01, 0.05, 0.1, 0.2, 0.5, 0.75, 1, 5, 10, 20, 50, 100]
}

#grid search k=5 fold cv
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, scoring='roc_auc', cv=5)
grid_search.fit(imputed_aps_train_X, aps_train_Y_encoded)
print(f"Best parameters found: {grid_search.best_params_}")
best_alpha = grid_search.best_params_['reg_alpha']
best_cv_error = 1 - grid_search.best_score_
best_xgb = XGBClassifier(objective='binary:logistic', reg_alpha=best_alpha, learning_rate=0.1, max_depth=5, use_label_encoder=False, eval_metric='logloss')
best_xgb.fit(imputed_aps_train_X, aps_train_Y_encoded)

train_probs = best_xgb.predict_proba(imputed_aps_train_X)[:, 1]
test_probs = best_xgb.predict_proba(imputed_aps_test_X)[:, 1]

train_fpr, train_tpr, _ = roc_curve(aps_train_Y_encoded, train_probs)
test_fpr, test_tpr, _ = roc_curve(aps_test_Y_encoded, test_probs)
train_roc_auc = auc(train_fpr, train_tpr)
test_roc_auc = auc(test_fpr, test_tpr)

#errs
train_err = 1 - train_roc_auc
test_err = 1 - test_roc_auc
print(f"CV Error for best alpha: {best_cv_error:.4f}")
print(f"Train Error: {train_err:.4f}")
print(f"Test Error: {test_err:.4f}")

#confusion matrix
train_preds = best_xgb.predict(imputed_aps_train_X)
test_preds = best_xgb.predict(imputed_aps_test_X)
train_confusion_matrix = confusion_matrix(aps_train_Y_encoded, train_preds)
test_confusion_matrix = confusion_matrix(aps_test_Y_encoded, test_preds)

print(f"Training Confusion Matrix:\n{train_confusion_matrix}")
print(f"Test Confusion Matrix:\n{test_confusion_matrix}")

#roc and auc
plt.plot(train_fpr, train_tpr, color='blue', lw=2,label=f'Train AUC = {train_roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--', color='gray', lw=2, linestyle='--')
plt.title('Training ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()

plt.plot(test_fpr, test_tpr, color='orange', lw=2,label=f'Test AUC = {test_roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--', color='gray', lw=2, linestyle='--')
plt.title('Test ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')

plt.show()
```

**The error of cross validation is larger than both training error and test error, this is because during cross validation we only use K-1 folds to train data, which result in higher bias.**

### (f) Use SMOTE to pre-process your data

```python
label_encoder = LabelEncoder()
aps_train_Y_encoded = label_encoder.fit_transform(aps_train_Y)
aps_test_Y_encoded = label_encoder.transform(aps_test_Y)

xgb = XGBClassifier(objective='binary:logistic', learning_rate=0.1, max_depth=5, use_label_encoder=False, eval_metric='logloss')

# right way to do cv, run smote for each k-1 folds
pipeline = IMBPipeline([
    ('smote', SMOTE(random_state=42)),
    ('classifier', xgb)
])

param_grid = {
    'classifier__reg_alpha': [0.01, 0.05, 0.1, 0.2, 0.5, 0.75, 1, 5, 10, 20, 50, 100]
}

grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, scoring='roc_auc', cv=5)
grid_search.fit(imputed_aps_train_X, aps_train_Y_encoded)

print(f"Best parameters: {grid_search.best_params_}")
best_alpha = grid_search.best_params_['classifier__reg_alpha']

xgb.set_params(**{'reg_alpha': best_alpha})

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(imputed_aps_train_X, aps_train_Y_encoded)

xgb.fit(X_resampled, y_resampled)

train_probs = xgb.predict_proba(X_resampled)[:, 1]
train_preds = xgb.predict(X_resampled)
test_probs = xgb.predict_proba(imputed_aps_test_X)[:, 1]
test_preds = xgb.predict(imputed_aps_test_X)

train_fpr, train_tpr, _ = roc_curve(y_resampled, train_probs)
test_fpr, test_tpr, _ = roc_curve(aps_test_Y_encoded, test_probs)
train_roc_auc = auc(train_fpr, train_tpr)
test_roc_auc = auc(test_fpr, test_tpr)

train_err = np.mean(train_preds != y_resampled)
test_err = np.mean(test_preds != aps_test_Y_encoded)

print(f"Training error: {train_err}")
print(f"Test error: {test_err}")

plt.plot(train_fpr, train_tpr, color='blue', lw=2, label=f'Train AUC = {train_roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--', color='gray', lw=2, linestyle='--')
plt.title('Training ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
plt.plot(test_fpr, test_tpr, color='orange', lw=2, label=f'Test AUC = {test_roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--', color='gray', lw=2, linestyle='--')
plt.title('Test ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
train_cm = confusion_matrix(y_resampled, train_preds)
test_cm = confusion_matrix(aps_test_Y_encoded, test_preds)
print(f"Training confusion matrix:\n{train_cm}")
print(f"Test confusion matrix:\n{test_cm}")
```

**The uncompensated model has a better training error as well as test error then the model using SMOTE.**

## 5. ISLR 9.7.3

```python
data_df = pd.DataFrame({
    'x1': [3, 2, 4, 1, 2, 4, 4],
    'x2': [4, 2, 4, 4, 1, 3, 1],
    'y': ['Red', 'Red', 'Red', 'Red', 'Blue', 'Blue', 'Blue']
})
data_df['y'].replace({'Red': 0, 'Blue': 1}, inplace=True)
input_data = data_df[['x1', 'x2']]
labels = data_df['y']

svm_model = svm.SVC(kernel='linear', C=1E10)
svm_model.fit(input_data, labels)

plt.figure(figsize=(7, 7))
plot_colors = ['red' if lbl == 0 else 'blue' for lbl in labels]
plt.scatter(input_data['x1'], input_data['x2'], c=plot_colors, s=50, edgecolors='k')
current_ax = plt.gca()
x_range = current_ax.get_xlim()
y_range = current_ax.get_ylim()
x_vals = np.linspace(x_range[0], x_range[1])
y_vals = np.linspace(y_range[0], y_range[1])
grid_y, grid_x = np.meshgrid(y_vals, x_vals)
grid_points = np.vstack([grid_x.ravel(), grid_y.ravel()]).T
boundary_values = svm_model.decision_function(grid_points).reshape(grid_x.shape)
b_0 = svm_model.intercept_[0]
b_1, b_2 = svm_model.coef_[0]
print(f"Hyperplane equation: {b_0:.2f} + {b_1:.2f}x1 + {b_2:.2f}x2 = 0")
plt.contour(grid_x, grid_y, boundary_values, colors='k', levels=[-1, 0, 1], linestyles=['--', '-', '--'])
plt.scatter(svm_model.support_vectors_[:, 0], svm_model.support_vectors_[:, 1], s=150, linewidth=1, facecolors='none', edgecolors='k', marker='o')
non_optimal_boundary = boundary_values + 0.1
plt.contour(grid_x, grid_y, non_optimal_boundary, colors='y', levels=[0], linestyles=['-'])
new_x1, new_x2 = 2.0, 3.5
plt.scatter(new_x1, new_x2, color='blue', marker='*')
```

### (a) The sketch is above

### (b) The solid black line is the hyperplane. Its equation is -1.00 + 2.00x1 + -2.00x2 = 0

### (c) If -1.00 + 2.00x1 + -2.00x2 >= 0, classified as blue, otherwise classified as red.

### (d) Margin is the 2 dashed black line.

### (e) Support vectors are circled points.

### (f) The 7th data point (4,1) is not a support vector and will not affect the hyperplane.

### (g) The yellow solid line can be a separating hyperplane but not the optimal one.      The equation is -0.9 + 2.00x1 + -2.00x2 = 0.

### (h) The blue star can be an additional point that make two classes are no longer separable by hyperplane.
